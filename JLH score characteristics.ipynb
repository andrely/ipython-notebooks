```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
```

The default relevance score for the significant terms aggregation is not given in the documentation. A quick dive into the code however and we find the following scoring function.

$$
JLH = \left\{\begin{matrix}
 (p_{fore} - p_{back})\frac{p_{fore}}{p_{back}} & p_{fore} - p_{back} > 0 \\ 
0  & elsewhere
\end{matrix}\right.
$$

Here the $p_{fore}$ is the frequency of the term in the foreground (or query) document set, while $p_{back}$ is the term frequency in the background document set which by default is the whole index.

Expanding the formula gives us the following which is quadratic in $p_{fore}$.

$$
(p_{fore} - p_{back})\frac{p_{fore}}{p_{back}} = \frac{p_{fore}^2}{p_{back}} - p_{fore}
$$

By keeping $p_{back}$ fixed and keeping in mind that both it and $p_{fore}$ is positive we get the following function plot. Note that $p_{back}$ is unnaturally large for illustration purposes.

```python
p_b = .5
p_f = np.linspace(0, 1)
plot(p_f, ((p_f**2)/p_b) - p_f, color='r', label=r'JLH score for $p_{back} = .5$')
plot([0,1], [0,0], ls='--', lw=2, color='black')
plt.legend()
plt.xlabel(r'$p_{fore}$')
plt.ylabel(r'JLH')
```

On the face of ot this looks bad for a scoring function. It can be undesirable that it changes sign, but more trouble some id the fact that this function is not monotonnically increasing.

The gradient of the function:

$$
\nabla JLH(p_{fore}, p_{back}) = \left(\frac{2 p_{fore}}{p_{back} - 1} , -\frac{p_{fore}^2}{p_{back}^2}\right)
$$

Setting the gradient to zero we see by looking at the second coordinate that the JLH does not have a minimum, but approaches it when $p_{fore}$ and $p_{back}$ approaches zero where the function is undefined. While the second coordinate is always positive, the first coordinate shows us where the function is not increasing.

$$
\begin{split}
\frac{2 p_{fore}}{p_{back}} - 1 & < 0 \\
p_{fore} & < \frac{1}{2}p_{back}
\end{split}
$$

Furtunately the decreasing part of the function is in an area where $p_{fore} - p_{back} < 0$ and the JLH score explicitly defined as zero. By symmetry of the square around the minimum of the first coordinate of the gradient around $\frac{1}{2}p_{back}$ we also see that the entire area where the score is below zero is in this region.

With this it seems sensible to just drop the linear term of the JLH score and just use the quadratic part. This will result in the same ranking with a slightly less steep increase in score as $p_{fore}$ increases.

$$
JLH_{mod} = \frac{p_{fore}^2}{p_{back}} 
$$

Looking at the level sets for the JLH score there is a quadratic relationship between the $p_{fore}$ and $p_{back}$. Solving for a fixed level $k$ we get:

$$
\begin{split}
JLH = & \frac{p_{fore}^2}{p_{back}} - p_{fore} = k \\
 & p_{fore}^2 - p_{fore} - k\cdot p_{back}  = 0 \\
 & p_{fore} = \frac{1}{2} \pm \frac{\sqrt{1 + 4 \cdot k \cdot p_{back}}}{2}
\end{split}
$$

Where the negative part is outside of function definition area.

This is far easier to see in the simplified formula.

$$
\begin{split}
JLH = & \frac{p_{fore}^2}{p_{back}} = k \\
 & p_{fore} = \sqrt{k \cdot p_{back}}
\end{split}
$$

An increase in $p_{back}$ must be offset by approximately a square root increase in $p_{fore}$ to  retain the same score.

```python
pf = np.linspace(0, 1, 200)
pb = np.linspace(0, 1, 200)
PF, PB = np.meshgrid(pf, pb)
JLH = (PF - PB)*(PF/PB)
contour(PF, PB, JLH, levels=np.linspace(0,10,100))
plt.xlabel(r'$p_{fore}$')
plt.ylabel(r'$p_{back}$')
```

As we see the score increases sharply as $p_{fore}$ increases in a quadratic manner against $p_{back}$. As $p_{back}$ becomes small compared to $p_{fore}$ the growth goes from linear in $p_{fore}$ to squared.

Finally a 3D plot of the score function.

```python
for i in xrange(len(JLH)):
    for j in xrange(len(JLH)):
        if pf[i] > pb[j]:
            JLH[i, j] = 0
            
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(PF, PB, JLH)
ax.set_xlabel(r'$p_{fore}$')
ax.set_ylabel(r'$p_{back}$')
ax.set_zlabel('JLH')
ax.set_zlim([0, 200])
plt.gca().invert_xaxis()
plt.gca().invert_yaxis()
plt.show()
```
